```python
data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'],axis=1,inplace=True)
data
```

```python
data.isnull().sum()
```

```python
data['Age']=data['Age'].fillna(data['Age'].median())
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isnull().sum()
```

```python
sb.countplot(x="Survived",data=data)
plt.title("Survived Count")
plt.show()
```

**Observation**
Most of the people not survived

```python
sb.countplot(x="Pclass", hue="Survived", data=data)
plt.title("Pclass vs Survived")
plt.show()
```

**Observation**
Passengers in the first class having the highest Survival rate tha 3rd class

```python
sb.boxplot(x="Survived", y="Age", data=data)
plt.title("Age vs Survival")
```

```python
sb.countplot(x="Sex", hue="Survived", data=data,palette="viridis")
plt.show()
```

**Observations**
survival rate is high for Females

```python
sb.lineplot(x="Pclass", y="Fare", data=data)
plt.title("Pclass vs Fare")
```

**Observation**
1st class passengers paid the highest fares, 2nd class paid less, and 3rd class paid the least.

```python
corr=data.select_dtypes(include=['int64','float64']).corr()
corr

sb.heatmap(corr,annot=True,cmap="Blues")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Sex']=le.fit_transform(data['Sex'])
data['Embarked']=le.fit_transform(data['Embarked'])

data.head()
```

```python
X = data.drop(['Sex','Embarked'],axis=1)
Y = data[['Sex','Embarked']]
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
X=ms.fit_transform(X)
y=pd.DataFrame(X,columns=[ 'Survived', 'Pclass',  'Age', 'SibSp',
       'Parch',  'Fare'])
y
```

```python
newdata=pd.concat([Y,y],axis=1)
newdata
```

```python
newdata.describe()
```

```python
sb.heatmap(newdata.corr(),annot=True)
plt.show()
```

**Summary**

1.   Main Features:-Sex, Pclass, Age, Fare
2.   Missing values:- Age & embarked had missing values and filled with the median and mode respectively
3.   Dropped the unimportant columns - 'PassengerId', 'Name', 'Ticket', 'Cabin'
2.   Gender, Class and age are the strongest predictors
